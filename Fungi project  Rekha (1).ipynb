# Encoding

```python
# Print the ratio of fraud cases
ratio = occurring/len(mushroom_df.index)
print(f'The Ratio of poisones mushroom is:{ratio[1]}\nThe Ratio of edible mushroom  is:{ratio[0]}')
```

```python
#separating the data of poisones and edible for analysis

Poisones_mushrooms = mushroom_df[mushroom_df['class'] == 1]
edible_mushrooms = mushroom_df[mushroom_df['class'] == 0]

print(Poisones_mushrooms.shape)
print(edible_mushrooms.shape)
```

```python
pois_sample=Poisones_mushrooms.sample(n=2092937)
```

```python
mushroom_df=pd.concat([pois_sample,edible_mushrooms],axis=0)
```

```python
mushroom_df.head(2)
```

```python
mushroom_df['class'].value_counts()
```

# Splitting  data into features and target

```python
# Copy all the independent variables into X dataframe
X = mushroom_df.drop('class', axis=1)

# Copy target into the y dataframe. 
y =mushroom_df[['class']]
```

```python
# Split X and y into training and test set in 70:30 ratio
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20 , random_state=2)
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler() 

X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

# Model building and Model Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.ensemble import RandomForestClassifier 
```

```python
import warnings
warnings.simplefilter("ignore")
```

# LogisticRegression

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

acc_lr = accuracy_score(y_test, y_pred_lr)
conf = confusion_matrix(y_test, y_pred_lr)
clf_report = classification_report(y_test, y_pred_lr)

print(f"Accuracy Score of Logistic Regression is : {acc_lr}")
print(f"Confusion Matrix : \n{conf}")
print(f"Classification Report : \n{clf_report}")
```

# Random Forest Classifier

```python
rd_clf = RandomForestClassifier()
rd_clf.fit(X_train, y_train)

y_pred_rd_clf = rd_clf.predict(X_test)

acc_rd_clf = accuracy_score(y_test, y_pred_rd_clf)
conf = confusion_matrix(y_test, y_pred_rd_clf)
clf_report = classification_report(y_test, y_pred_rd_clf)

print(f"Accuracy Score of Random Forest is : {acc_rd_clf}")
print(f"Confusion Matrix : \n{conf}")
print(f"Classification Report : \n{clf_report}")
```
